## Import the necessary libraries

```python
import pandas as pd
```

## Assign it to a variable called users.

```python
users= pd.read_csv('CardioGoodFitness.csv')
users.head()
```

## Present general statistics by Product

```python
users.groupby('Product').describe()
```

## For each combination of Product and Gender, calculate the mean of Age, Fitness, and Income.

```python
users.groupby(['Product','Gender'])['Age','Fitness','Income'].mean()
```

```python
users.groupby(['Product','Gender'], as_index=False)['Age','Fitness','Income'].mean()
```

## Discover the Male ratio per Product and sort it from the most to the least

```python
# create a function
def gender_to_numeric(x):
    if x == 'Male':
        return 1
    if x == 'Female':
        return 0
# apply the function to the gender column and create a new column
users['gender_n'] = users['Gender'].apply(gender_to_numeric)  

a = users.groupby('Product').gender_n.sum() / users.Product.value_counts() * 100 
# sort to the most male 
a.sort_values(ascending = False)
```

## For each Product, calculate the minimum and maximum ages

```python
users.groupby('Product').Age.agg(['min', 'max'])
```

## For each Product present the percentage of women and men

```python
# create a data frame and apply count to gender
gender_ocup = users.groupby(['Product', 'Gender']).agg({'Gender': 'count'})

# create a DataFrame and apply count for each occupation
occup_count = users.groupby(['Product']).agg('count')

# divide the gender_ocup per the occup_count and multiply per 100
occup_gender = gender_ocup.div(occup_count, level = 'Product') * 100

# present all rows from the 'gender column'
occup_gender.loc[: , 'Gender']
```

## What is the number of observations in each Product and MaritalStatus

```python
users.groupby(['Product', 'MaritalStatus']).size()
```
